# Part A

### Generate random obstacles with generateMap2d and find shortest path to goal with different search algorithms.

```python
import path_planning as pp
import matplotlib.pyplot as plt
import numpy as np

"""
create maps with obstacles randomly distributed
cells with a value of 0: Free cell; 
                     -1: Obstacle;
                     -2: Start point;
                     -3: Goal point;
"""

# create a map with obstacles randomly distributed
_map_ = pp.generateMap2d([60,60])
plt.clf()
plt.title('Generated map with random obstacles')
plt.imshow(_map_)
plt.savefig('regular_map2d.png')
start, goal = pp.findStartGoal(_map_)
print('Start: ' + str(start))
print('Goal: ' + str(goal))
```

## Uninformed search

### Random search

```python
import random_search as rs

path, solved_map = rs.search(_map_, start, goal)
pp.plotMap(solved_map, path, title_='Solved path with Random Search', filename_='path_random_search.png')
```

### Breadth-first search

```python
import bfs_search as bfs 

path, solved_map = bfs.search(_map_, start, goal)
pp.plotMap(solved_map, path, title_='Solved path with with BFS', filename_='path_breadth_first_search.png')
```

### Depth-first search

```python
import dfs_search as dfs

path, solved_map = dfs.search(_map_, start, goal)
pp.plotMap(solved_map, path, title_='Solved path with with DFS', filename_='path_depth_first_search.png')
```

## Informed search

### Greedy search - euclidean

```python
import greedy_search as gs

path, solved_map = gs.search(_map_, start, goal, metric='euclidean')
pp.plotMap(solved_map, path, title_='Solved path with Greedy search (euclidean)', filename_='path_greedy_search_euclidean.png')
```

### Greedy search - manhattan

```python
path, solved_map = gs.search(_map_, start, goal, metric='manhattan')
pp.plotMap(solved_map, path, title_='Solved path with Greedy search (manhattan)', filename_='path_greedy_search_manhattan.png')
```

### A* search - euclidean

```python
import A_star
path, solved_map, exp_nodes = A_star.search(_map_, start, goal, metric='euclidean')
pp.plotMap(solved_map, path, title_='Solved path with A* search (euclidean)', filename_='path_a_star_euclidean.png')
```

### A* search - manhattan

```python
import A_star
path, solved_map, exp_nodes = A_star.search(_map_, start, goal, metric='manhattan')
pp.plotMap(solved_map, path, title_='Solved path with A* search (manhattan)', filename_='path_a_star_manhattan.png')
```

# Part B

### Generate map with predefined obstacles and find shortest path with different search algorithms.

```python
# create a map with rotated H shape obstacle and obstacles randomly distributed
map_h_object, info = pp.generateMap2d_obstacle([60,60])
start, goal = pp.findStartGoal(map_h_object)
print('Start: ' + str(start))
print('Goal: ' + str(goal)) 
plt.clf()
plt.title('Generated map with H-shape obstacle')
plt.imshow(map_h_object)
plt.savefig('h_obstacle_map.png')
```

### Random search

```python
import random_search as rs
path, solved_map = rs.search(map_h_object, start, goal)
pp.plotMap(solved_map, path, title_='Solved path with Random search', filename_='h_path_random_search.png')
```

### Breadth-first search

```python
import bfs_search as bfs
path, solved_map = bfs.search(map_h_object, start, goal)
pp.plotMap(solved_map, path, title_='Solved path with BFS', filename_='h_path_breadth_first_search.png')
```

### Depth-first search

```python
import dfs_search as dfs
path, solved_map = dfs.search(map_h_object, start, goal)
pp.plotMap(solved_map, path, title_='Solved path with DFS', filename_='h_path_depth_first_search.png')
```

### Greedy search - euclidean

```python
path, solved_map = gs.search(map_h_object, start, goal, metric='euclidean')
pp.plotMap(solved_map, path, title_='Solved path with Greedy search (euclidean)', filename_='h_path_greedy_euclidean.png')
```

### Greedy search - manhattan

```python
path, solved_map = gs.search(map_h_object, start, goal, metric='manhattan')
pp.plotMap(solved_map, path, title_='Solved path with Greedy search (manhattan)', filename_='h_path_greedy_manhattan')
```

### A* search - euclidean

```python
import A_star
path, solved_map, exp_nodes = A_star.search(map_h_object, start, goal, metric='euclidean')
pp.plotMap(solved_map, path, title_='Solved path with A* search (euclidean)', filename_='h_path_a_star_euclidean.png')
print(f'Expanded nodes: {exp_nodes}')
```

### A* search - manhattan

```python
import A_star
path, solved_map, exp_nodes = A_star.search(map_h_object, start, goal, metric='manhattan')
pp.plotMap(solved_map, path, title_='Solved path with A* search (manhattan)', filename_='h_path_a_star_manhattan.png')
print(f'Expanded nodes: {exp_nodes}')
```

### A* search - special heuristic

```python
import A_star
path, solved_map, exp_nodes = A_star.search(map_h_object, start, goal, metric='special', info=info)
pp.plotMap(solved_map, path, title_='Solved path with A* search (special)', filename_='h_path_a_star_special.png')
print(f'Expanded nodes: {exp_nodes}')
```
