# Part A

### Generate random obstacles with generateMap2d and find shortest path to goal with different search algorithms.

```python
import path_planning as pp
import matplotlib.pyplot as plt
import numpy as np

"""
create maps with obstacles randomly distributed
cells with a value of 0: Free cell; 
                     -1: Obstacle;
                     -2: Start point;
                     -3: Goal point;
"""

# create a map with obstacles randomly distributed
_map_ = pp.generateMap2d([60,60])
plt.clf()
plt.title('Generated map')
plt.imshow(_map_)
start, goal = pp.findStartGoal(_map_)
print('Start: ' + str(start))
print('Goal: ' + str(goal))
```

## Uninformed search

### Random search

```python
import random_search as rs

path, solved_map = rs.search(_map_, start, goal)
pp.plotMap(solved_map, path, title_='Solved path Random Search')
```

### Breadth-first search

```python
import bfs_search as bfs 

path, solved_map = bfs.search(_map_, start, goal)
pp.plotMap(solved_map, path, title_='Solved path BFS')
```

### Depth-first search

```python
import dfs_search as dfs

path, solved_map = dfs.search(_map_, start, goal)
pp.plotMap(solved_map, path, title_='Solved path DFS')
```

## Informed search

### Greedy search - euclidean

```python
import greedy_search as gs

path, solved_map = gs.search(_map_, start, goal, metric='euclidean')
pp.plotMap(solved_map, path, title_='Solved path Greedy search (euclidean)')
```

### Greedy search - manhattan

```python
path, solved_map = gs.search(_map_, start, goal, metric='manhattan')
pp.plotMap(solved_map, path, title_='Solved path Greedy search (manhattan)')
```

### A* search - euclidean

```python
import A_star
path, solved_map = A_star.search(_map_, start, goal, metric='euclidean')
pp.plotMap(solved_map, path, title_='Solved path A* search (euclidean)')
```

### A* search - manhattan

```python
import A_star
path, solved_map = A_star.search(_map_, start, goal, metric='manhattan')
pp.plotMap(solved_map, path, title_='Solved path A* search (manhattan)')
```
